```python
import os
import json
import re
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
ROOT = 'C:\\Users\\niti.mishra\\Documents\\Personal\\cyberbullying\\'
RESULTS = Path(os.path.join(ROOT, 'results\\1g2g_learningcurve'))
PLOTS = Path(os.path.join(RESULTS))
files = [i for i in list(RESULTS.iterdir()) if '.json' in i.suffix ]
```

```python
file = files[0]
df = pd.read_json(file, lines=True)
df
```

```python
full_df = pd.DataFrame()
for i in files:
    each_df = pd.read_json(i, lines=True)
    each_df['iter'] =  re.findall(r'\d+', i.stem)*len(each_df)
    each_df['iter'] = each_df['iter'].astype(int)
    full_df = full_df.append(each_df)

full_df.head()
```

```python
def avg_series(series):
    return [sum(i)/len(i) for i in series]
full_df.iloc[:,2:7] = full_df.iloc[:,2:7].apply(avg_series)
# full_df.to_excel("output.xlsx")
```

```python
plot_df = full_df.groupby(['name', 'iter'])['f1'].sum().unstack('name').reset_index()
plot_df = plot_df.loc[:,['NB' not in i for i in plot_df.columns]]
plot_df
```
