```python
import os
import json
import re
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import ast
```

```python
PATH = '/Users/peaceforlives/Documents/Projects/cyberbullying/final_classification/'
```

```python
file = open(PATH+'tfidf_dictionary.txt', 'r')
contents = file.read()
vocab = ast.literal_eval(contents)
file. close()
# print(dictionary)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15,5))
lists_asc = sorted(vocab.items())
x = [i for (i,j) in lists_asc]
y = [j for (i,j) in lists_asc]

n=30
plt.bar(x[2:n], y[2:n])
plt.xticks(rotation=45)
plt.show()
```

```python
from matplotlib.font_manager import FontProperties

prop = FontProperties(fname='/System/Library/Fonts/Apple Color Emoji.ttc')
plt.rcParams['font.family'] = prop.get_family()

fig, ax = plt.subplots(1, 1, figsize=(15,5))
lists_asc = sorted(vocab.items(), reverse=True)
x = [i for (i,j) in lists_asc]
y = [j for (i,j) in lists_asc]

n=30
plt.bar(x[:n], y[:n])
plt.xticks(rotation=45)
ax.set_xticklabels(x, rotation=0, fontsize=5, fontproperties=prop)
plt.show()
```
